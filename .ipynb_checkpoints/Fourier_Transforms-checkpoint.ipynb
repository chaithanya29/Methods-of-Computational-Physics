```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import scipy.signal as ss
```

## DFT

```python
def DFT(ar):
    n = len(ar)
    dft = np.array([])
    for i in range(n):
        b=0
        for j in range(n):
            b = b + (ar[j]*np.exp(-2.j*(np.pi*i*j)/n))
        dft = np.append(dft,b)
        
    return(dft)
```

### Sine wave

```python
Fs = 150 #sampling frequqncy
t = np.linspace(0,1,Fs) #time vector
freq = 5 # frequency
nfft = 1024 #length of fft
x_sin = np.sin(2*np.pi*t*freq) #sine wave of frequency f Hz
x_sin_fft = abs(np.fft.fft(x_sin,nfft)[1:nfft//2]) #performing fft and taking the absolute of its one symmetric part
f = np.linspace(0,1,nfft/2-1)*Fs/nfft # frequency vector
```

```python
plt.plot(t,x_sin)
plt.title("Sine wave signal")
plt.xlabel("Time(s)")
plt.ylabel("Amplitude")
plt.show()
```

```python
plt.plot(f,x_sin_fft)
plt.title("Power spectrum of sine wave")
plt.xlabel("Time(s)")
plt.ylabel("Power")
plt.show()
```

### Square pulse

```python
Fs = 150 #sampling frequqncy
t = np.linspace(-0.5,0.5,Fs) #time vector
w = .2 # width of rectangle
nfft = 512 #length of fft
x_sq = ss.square(t,w) #square wave of width 0.2
x_sq_fft = abs(np.fft.fft(x_sq,nfft)[1:nfft//2]) #performing fft and taking the absolute of its one symmetric part
f = np.linspace(0,1,nfft/2-1)*Fs/nfft # frequency vector
```

```python
plt.plot(t, x_sq)
plt.title("Square Pulse signal")
plt.xlabel("Time(s)")
plt.ylabel("Amplitude")
plt.show()
```

```python
plt.plot(f, x_sq_fft)
plt.title("Power spectrum of Square Pulse signal")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Power")
plt.show()
```

### Gaussian pulse

```python
Fs = 60 #sampling frequqncy
t = np.linspace(-.5,0.5,Fs)
nfft = 1024 #length of fft
x_g = 1/(np.sqrt(1*np.pi*0.01))*(np.exp(-(t**2)/(2*0.01))) #Gaussian signal of std 0.01
x_g_fft = abs(np.fft.fft(x_g,nfft)[1:nfft//2]) #performing fft and taking the absolute of its one symmetric part
f = np.linspace(0,1,nfft/2-1)*Fs/nfft # frequency vector
```

```python
plt.plot(t, x_g)
plt.title("Gaussian Pulse signal")
plt.xlabel("Time(s)")
plt.ylabel("Amplitude")
plt.show()
```

```python
plt.plot(f, x_g_fft)
plt.title("Power spectrum of Gaussian Pulse signal")
plt.xlabel("Frequency")
plt.ylabel("Power")
plt.show()
```

### Exponential Decay Signal

```python
Fs = 150 #sampling frequqncy
t = np.linspace(0,1,Fs)
nfft = 1024 #length of fft
x_exp = 2*np.exp(-5*t) #exponential signal
x_exp_fft = abs(np.fft.fft(x_exp,nfft)[1:nfft//2]) #performing fft and taking the absolute of its one symmetric part
f = np.linspace(0,1,nfft/2-1)*Fs/nfft # frequency vector
```

```python
plt.plot(t, x_exp)
plt.title("Exponential Decay Pulse signal")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.show()
```

```python
plt.plot(f, x_exp_fft)
plt.title("Power spectrum of exponential Pulse signal")
plt.xlabel("Frequency")
plt.ylabel("Power")
plt.show()
```

### Chirp Signal

```python
Fs = 200 #sampling frequqncy
t = np.linspace(0,1,Fs)
nfft = 1024 #length of fft
x_ch = ss.chirp(t,0,1,Fs/6) #chirp signal
x_ch_fft = abs(np.fft.fft(x_ch,nfft)[1:nfft//2]) #performing fft and taking the absolute of its one symmetric part
f = np.linspace(0,1,nfft/2-1)*Fs/nfft # frequency vector
```

```python
plt.plot(t, x_ch)
plt.title("Chirp Pulse signal")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.show()
```

```python
plt.plot(f, x_ch_fft)
plt.title("Power spectrum of exponential Pulse signal")
plt.xlabel("Frequency")
plt.ylabel("Power")
plt.show()
```

# Using Euler and Runge-Kutta methods to solve ODE's

```python
import numpy as np
```

```python
def f1(t,y):
    return (-t*y)
```

```python
def f2(t,y):
    return(t**2 + 1)
```

```python
def euler(t0,y0,h,f,n):
    y_ar = np.zeros(n+1)
    i = 0
    y_ar[0] = y0
    t = t0
    while i<n:
        y_ar[i+1] = y_ar[i] + h*f(t, y_ar[i])
        t += h
        i += 1
    return(y_ar)
```

```python
def mid_euler(t0,y0,h,f,n):
    y_ar = np.zeros(n+1)
    i = 0
    y_ar[0] = y0
    t = t0
    while i<n:
        y_half = y_ar[i]+(h/2)*f(t, y_ar[i])
        y_ar[i+1] = y_ar[i] + h*f(t+h/2, y_half)
        t += h
        i += 1
    return(y_ar)
```

```python
def average_method(t0,y0,h,f,n):
    y_ar = np.zeros(n+1)
    i = 0
    y_ar[0] = y0
    t = t0
    while i<n:
        y_ar[i+1] = y_ar[i] + (h/2)*(f(t, y_ar[i])+f(t+h,y_ar[i]+h*f(t,y_ar[i])))
        t += h
        i += 1
    return(y_ar)
```

```python
def runge_kutta(t0,y0,h,f,n):
    y_ar = np.zeros(n+1)
    i = 0
    y_ar[0] = y0
    t = t0
    while i<n:
        k0 = f(t,y_ar[i])
        k1 = f(t+h/2, y_ar[i]+(h/2)*k0)
        k2 = f(t+h/2, y_ar[i]+(h/2)*k1)
        k3 = f(t+h, y_ar[i]+h*k2)
        y_ar[i+1] = y_ar[i] + (h/6)*(k0 + 2*k1+ 2*k2 + k3)
        t += h
        i += 1
    return(y_ar)
```

```python
def plot(N, h, f, fy, y0):
    plt.plot(np.linspace(0,5,N),runge_kutta(0, y0, h, f, N-1))
    plt.plot(np.linspace(0,5,N),euler(0, y0, h, f, N-1))
    plt.plot(np.linspace(0,5,N),mid_euler(0, y0, h, f, N-1))
    plt.plot(np.linspace(0,5,N),average_method(0, y0, h, f, N-1))
    plt.plot(np.linspace(0,5,N),fy(np.linspace(0,5,N)))
    plt.legend(['Runge-Kutta','Euler','Mid-Point','Average','Original Analytic solution'])
```

```python
def shm_euler(t0, y, K ,tf, N): # SHM using improved (Mid-Point) Euler method
    y1_ar = np.zeros(N+1)
    y2_ar = np.zeros(N+1)
    y1_ar[0] = y[0]
    y2_ar[0] = y[1]

    h = (tf-t0)/N
    #t = t0
    for i in range(N):
        #k = K(t,y1_ar[i], y2_ar[i])
        #y1_ar[i+1] = y1_ar[i] + k[0]*h
        #y2_ar[i+1] = y2_ar[i] + k[1]*h
        k_half = K(y1_ar[i], y2_ar[i])
        y1_half = y1_ar[i]+(h/2)*k_half[0]
        y2_half = y2_ar[i]+(h/2)*k_half[1]
        k = K(y1_half, y2_half)
        y1_ar[i+1] = y1_ar[i] + k[0]*h
        y2_ar[i+1] = y2_ar[i] + k[1]*h
        #t = t+h
    return(y1_ar,y2_ar)
```

```python
def shm_rk(t0, y, K ,tf, N): # SHM using Runge-Kutta method
    y1_ar = np.zeros(N+1)
    y2_ar = np.zeros(N+1)
    y1_ar[0] = y[0]
    y2_ar[0] = y[1]

    h = (tf-t0)/N
    t = t0
    for i in range(N):
        
        k0 = K(y1_ar[i], y2_ar[i])
        k1 = K(y1_ar[i]+(h/2)*k0[0], y2_ar[i]+(h/2)*k0[1])
        k2 = K(y1_ar[i]+(h/2)*k1[0],y2_ar[i]+(h/2)*k1[1])
        k3 = K(y1_ar[i]+h*k2[0],y2_ar[i]+h*k2[1])
        y1_ar[i+1] = y1_ar[i] + (h/6)*(k0[0] + 2*k1[0]+ 2*k2[0] + k3[0])
        y2_ar[i+1] = y2_ar[i] + (h/6)*(k0[1] + 2*k1[1]+ 2*k2[1] + k3[1])
        t = t+h
    return(y1_ar,y2_ar)
```

```python
def K_shm(y1, y2):
    k = 1
    m = 1
    return([y2,-(k/m)*y1])
```

```python
y_1, y_2 = shm_euler(0,[1,0],K_shm,50,100)
```

```python
def plot_shm(N,shm):
    y_1, y_2 = shm(0,[1,0],K_shm,50,N)
    plt.plot(y_1)
    plt.plot(y_2)
    plt.ylim(-5,5)
    plt.legend(['x','v'])
```

```python
plot_shm(50, shm_euler) # h= 50
```

```python
plot_shm(1000, shm_rk) # h = 1000
```

As we can see above, the decrease in step size significantly reduces the amplification of the error and gives reasonable plots. The Runge_Kutta method performs much better than the Euler method

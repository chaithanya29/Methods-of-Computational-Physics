```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def monte_carlo(N):
    c = 0
    for i in range(N):
        x = np.random.random()
        y = np.random.random()
        if x**2 + y**2 <=1:
            c +=1
    return(c)
```

```python
4*monte_carlo(1000000)/1000000
```

```python
X = np.linspace(10,1000000,400)
pi = np.array([])
for i in X:
    pi = np.append(pi,4*monte_carlo(int(i))/i)
```

```python
plt.plot(X[1:],pi[1:]) #excluded 1st point for better visualization
plt.plot(X[1:],[3.14]*399)
plt.xlabel('Total number of points')
plt.ylabel('Value of Pi estimated')
#plt.ylim((3,3.6))
plt.show()
```

```python
X1 = np.array([])
for i in range(100):
    X1 = np.append(X1,4*monte_carlo(100)/100)
```

```python
plt.hist(X1)
plt.xlabel('Bins of Pi values')
plt.ylabel("Number of occurances out of 100")
plt.show()
```
